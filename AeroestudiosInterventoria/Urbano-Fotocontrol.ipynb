### *Verificación Shapefile*

```python
#verificación Shapefile
fshp=fmun+folder+"\\1. Archivo geografico"

for filename in glob.iglob(fshp+"\\*.shp"):
    print(filename)
    shp=gpd.read_file(filename)
    name=filename.split('\\1. Archivo geografico\\')
    print("----------> "+name[1]+" puntos: "+str(shp.geometry.count()))
    puntos=shp[["NAME","ESTE","NORTE","Altura_Ort"]]; print(puntos)
```
